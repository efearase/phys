# Вопрос по выбору:
## Моделирование дифракционной решетки.

Дифракционная решетка — важнейший спектральный прибор, предназначенный для разложения света в спектр и измерения
длин волн. Она представляет собой плоскую стеклянную или металлическую поверхность, на которой делительной машиной нарезано очень много (до сотен тысяч) прямых равноотстоящих штрихов.
Сначала рассмотрим простейшую идеализированную решетку, состоящую из одинаковых равноотстоящих параллельных щелей, сделанных в непрозрачном экране. Ширину щели обозначим через $b$, ширину непрозрачной‘ части экрана между двумя соседними щелями — через $a$. Величина $d = а - b$ называется периодом решетки.
В решетке осуществляется многолучевая интерференция когерентных дифрагированных пучков света, исходящих от щелей решетки при ее освещении. Дифракционная картина наблюдается по методу Фраунгофера, т.е. либо на бесконечно удаленном экране, либо в фокальной плоскости линзы, поставленной на пути дифрагированного света.
Пусть на решетку перпендикулярно к ее поверхности падает
плоская монохроматическая волна. 

<img src="1.png"/>

Разность хода между вторичными волнами, исходящими из соседних щелей решетки, будет $d \sin \theta $, a разность $\delta = kd \sin \theta$, где $\theta$ — угол дифракции. Обозначим через $E_1$, поле в точке наблюдения, излучаемое первой щелью. Оно определяется формулой $E_1 = b \frac{\sin \alpha }{\alpha}$. Поля, излучаемые остальными щелями, представятся выражениями:

$E_2 = E_1 \cdot e^{-i \delta}$ , $E_3 = E_2 \cdot e^{-i \delta} = E_1 \cdot e^{-i2 \delta}$    ...        $E_N = E_1 \cdot e^{-i(N-1) \delta}$,

где N - полное число щелей. Тогда полное поле, излучаемое всеми щелями представляется суммой:

$E = E_1 ( 1 + e^{-i \delta} + e^{-i 2 \delta} ...  + e^{-i( N-1)\delta})$ = $E_1 \frac{1 -e^{-iN\delta}}{1-e^{-i\delta}}$ = $ E_1 \frac{\sin (N\delta /2)} {\sin (\delta /2)} e^{i \phi}$ 

$A = A_1 \frac{1 -e^{-iN\delta}}{1-e^{-i\delta}}$ = $ E_1 \frac{\sin (N\delta /2)} {\sin (\delta /2)} $ (1), 

где $A_1$- вещественная амплитуда от одной щели, $A$- от всей решетки.

Тогда для интенсивности имеем:

$ I  = E \cdot E^* = E_1^2 \left(\frac{\sin (N\delta /2)} {\sin (\delta /2)}\right)^2 = I_1\left(\frac{\sin (N\delta /2)} {\sin (\delta /2)}\right)^2  $ (2)

Когда $\theta = 0$ , то $\delta = 0$. В этом случае (1) и (2) принимают неопределенный вид 0/0. Раскрыв неопределенность по правилу Лопиталя, получаем: 

$A_{гл}= NA_1$, $ I_{гл}= N^2 I_2$ (3)

Тот же результат получается при $\frac {\delta}{2} = m \pi$, те при

$d \sin \theta = m \lambda$,   $m \in Z$(4)

В направлениях, определяемых этим условием, получаются максимумы, интенсивность которых в $N^2$ раз превосходит интенсивность
волны от одной шели в том же направлении. Они называются главными максимумами. Целое число $m$ называют порядком главного максимума или порядком спектра. Условие (4) определяет направления, в которых излучения от всех щелей решетки приходят в точку наблюдения в одинаковых фазах, а потому усиливают друг друга. В таких направлениях при отдельных значениях $m$ могут и не возникнуть максимумы. Это будет, когда $I_1 = 0$, т. е. в направлениях на дифракционные минимумы от одной щели. Например, если $a = b$, то все главные максимумы четных порядков не появятся. Действительно, условие появления главного максимума порядка $2n$ имеет вид $d \sin \theta = 2n \lambda$, при $d= 2b$ оно переходит в $b \sin \theta = n \lambda$, т.е. в условие дифракционного минимума на щели. Таким образом, в рассматриваемом направлении ни одна шель, а потому и решетка в целом не излучают.

Выражения (1) и (2) обращаются в ноль, если $\sin (N \delta /2) = 0,$ но $\sin ( \delta /2) \ne 0$, те при  $N \delta /2 = (Nm+p)\pi$ или 

$d \sin \theta = (m+\frac{p}{N}) \lambda$, где $p = 1,2,...,N-1$
В соответствующих направлениях получаются дифракционные минимумы, в которых интенсивность света равна нулю. Между двумя соседними минимумами получается максимум. Такие максимумы называются второстепенными или добавочными. Между двумя соседними главными максимумами располагается (N — 1) минимумов и (N — 2) добавочных максимумов. На эти максимумы и минимумы накладываются минимумы, возникающие при дифракции
от отдельной щели, в которых функция $E_1$ обращается в нуль.
Второстепенные максимумы находятся примерно посередине
между соответствующими дифракционными минимумами. Величину $\delta$, определяющую направление на какой-либо из максимумов, можно вычислить по приближенной формуле:

$\frac{\delta}{2} = \left( m+ \frac{2p+1}{2N}\right) \pi$

### Дифракционная решетка, как спектральный прибор
Положение главных максимумов в дифракционной решетке зависит от длииы волны. Исключение составляют только главные максимумы нулевого порядка ($m = 0$), положения которых от длины волны не зависят. Белый и всякий сложный свет можно рассматривать как суперпозицию монохроматических волн с различными длинами. Эти волны при дифракции на решетке ведут себя независимо. Поэтому решетка в каждом порядке $m \ne 0$ разложит падающий свет в спектр, в котором отдельные монохроматические компоненты окажутся пространственно разделенными. Главные дифракционные максимумы, соответствующие $m= 1$, образуют спектр первого порядка. За ним идет спектр второго $m= 2$, третьего $m= 3$ и высших порядков. Если падающий свет белый, то спектр каждого порядка имеет вид цветной полосы, в которой встречаются все цвета радуги. В такой полосе наиболее отклоненными будут красные лучи, наименее отклоненными — фиолетовые. Положение спектральных линий в спектрах дифракционной
решетки определяется простыми соотношением (4). Спектр называется нормальным, если координата х, характеризующая положение спектральной линии в спектре, линейно меняется с длиной волны. При малых углах дифракции, когда изменением косинуса угла $\theta$ можно пренебречь, дифракционная решетка дает нормальный спектр.

Важными характеристиками дифракционной решетки и других спектральных аппаратов являются угловая дисперсия, дислерсионная область и разрешающая способность.

Угловой дисперсией называется производная $\frac{d \theta}{d\lambda}$. Чем больше угловая дисперсия, тем больше расстояние в спектре между двумя спектральными линиями с фиксированными длинами волн. Угловая дисперсия независит от параметров решетки.
Дисперсионная область. Если спектры соседних порядков перекрываются, то спектральный аппарат становится непригодным для исследования соответствующего участка спектра. Максимальная ширина спектрального интервала $ \Delta \lambda $, при которой еще нет перекрытия, называется дисперсионной областью спектрального аппарата. Найдем дисперсионную область для дифракционной решетки. Пусть длины волн падающего излучения лежат в спектральном интервале от $\lambda\ \ до \ \  \lambda’ = \lambda + \Delta \lambda$ . Пусть правый конец спектра
(m+1)-го порядка для длины волны $\lambda$ совпадает по своему поло-
жению с левым конпом спектра m-го порядка для длины волны $\lambda'$, тогда 

$\Delta \lambda = \lambda /m $

Это и есть дисперсионная область дифракционной решетки в рас-
сматриваемом участке спектра. При заданной длине волны она
определяется только порядком спектра m. Чем больше m, тем уже
дисперсионная область. В дифракционных решетках используются
спектры низких порядков (обычно второго или третьего). Поэтому
дифракционные решетки характеризуются широкими областями
дисперсии и пригодны для исследования широких участков спектра.
В этом основное преимущество дифракционных решеток перед
интерференционными спектральными аппаратами, у которых из-за
высоких порядков m дисперсионные области очень узкие.

Наименьшая
разность длин волн двух спектральных линий $\delta \lambda$, при которой спектральный аппарат разрешает эти линии, называется спектральным
разрешаемым расстоянием, а величина $R = \frac{\lambda}{\delta \lambda}$ — разрешающей
способностью аппарата.

Разрешаемое спектральное расстояние $\delta \lambda$ относится к числу 
не вполне точно определенных понятий и может быть указано лишь
ориентировочно. Для дифракционной решетки Рэлей предложил
следующий критерий спектрального разрешения. Спектральные
линии с близкими длинами волн $\lambda$ и $\lambda'$ считаются разрешенными,
если главный максимум дифракционной картины для одной длины
волны совпадает по своему положению с первым дифракционным
минимумом в том же порядке для другой длины волны. Если такой
критерий выполняется, то $R = \frac{\lambda}{\delta \lambda} = mN$

Рассмотрим распределение
интенсивности для спектральных линий $\lambda$ и $\lambda'$, сдвинутых одна
относительно другой на расстояние $\delta \lambda$. Так как спектральные линии с различными длинами волн некогерентны, то результирующая интенсивность найдется простым сложением интенсивностей
обеих линий. Она представлена на риcунке сплошной кривой, когда
интенсивности обеих спектральных линий одинаковы, а расстояние
между ними $\delta \lambda$ соответствует критерию Рэлея. В центре кривой
получился провал или минимум, интенсивность в котором, как легко
подсчитать, составляет около 80% максимальной. При наличии
такого провала нормальный глаз обычно воспринимает получившуюся картину как «двойную спектральную линию».

Для повышения разрешающей способности можно либо увеличивать число штрихов N, либо повьшнать порядок интерференции m. Первый путь используется в дифракционных решетках.
второй — в интерференционных спектральных приборах.

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import Slider
from IPython.display import Image


def grated_diffraction_intensity (slit_width, wavelength, screen_distance, distance_between_slits, number_of_slits, X):
  term1 = np.sin(np.pi*X*slit_width/(wavelength*screen_distance))/(np.pi*X*slit_width/(wavelength*screen_distance))
  term2 = (np.sin(number_of_slits*np.pi*distance_between_slits*X/(wavelength*screen_distance)))/(number_of_slits*np.sin((np.pi*distance_between_slits*X)/(wavelength*screen_distance)))
  return (term1**2)*(term2**2)

X = np.arange(-0.01,0.01,0.00001)
slit_width = 100*(10**-6)
wavelength1 = 495*(10**-9)
wavelength2 = 555*(10**-9)
wavelength3 = 760*(10**-9)
screen_distance = 95*(10**-2)
distance_between_slits= 1*10**-3
number_of_slits = 50

Y1 = grated_diffraction_intensity(slit_width, wavelength1, screen_distance, distance_between_slits, number_of_slits, X)
Y2 = grated_diffraction_intensity(slit_width, wavelength2, screen_distance, distance_between_slits, number_of_slits, X)
Y3 = grated_diffraction_intensity(slit_width, wavelength3, screen_distance, distance_between_slits, number_of_slits, X)
plot1, = plt.plot(X,Y1,color = 'blue')
plot2, = plt.plot(X,Y2,color = 'green')
plot3, = plt.plot(X,Y3,color = 'red')
plot, = plt.plot(X,Y1+Y2+Y3,color = 'black')
plt.xlabel("Distance from center")
plt.ylabel("Intensity")


axis1 = (plt.axes([0.75, 0.75, 0.14, 0.05]))
axis2 = (plt.axes([0.75,0.65, 0.14, 0.05]))
axis3 = (plt.axes([0.75,0.55, 0.14, 0.05]))
axis4 = (plt.axes([0.75,0.45, 0.14, 0.05]))
axis5 = (plt.axes([0.75,0.35, 0.14, 0.05]))
axis6 = (plt.axes([0.75,0.25, 0.14, 0.05]))
axis7 = (plt.axes([0.25,0.90, 0.65, 0.03]))

wavelength1_slider = Slider(axis1,'Wavelength 1(nm)',100, 1000,valinit=wavelength1*10**9 )
wavelength2_slider = Slider(axis2,'Wavelength 2(nm)',100, 1000,valinit=wavelength2*10**9 )
wavelength3_slider = Slider(axis3,'Wavelength 3(nm)',100, 1000,valinit=wavelength3*10**9 )
slit_width_slider = Slider(axis4, "Slit Width(mcm)", 10, 1000, valinit=slit_width*10**6)
screen_distance_slider = Slider(axis5, "Screen Distance(cm)", 10, 100, valinit= screen_distance*10**2)
distance_between_slits_slider = Slider(axis6, "Distance b/w slits(mm)", 0.1, 10, valinit=distance_between_slits*10**3)
number_of_slits_slider = Slider(axis7, "Number of slits", 1, 100, valinit=number_of_slits, valfmt='%0.0f')

def update(val):
  wavelength1 = wavelength1_slider.val*(10**-9)
  wavelength2 = wavelength2_slider.val*(10**-9)
  wavelength3 = wavelength3_slider.val*(10**-9)
  slit_width = slit_width_slider.val*(10**-6)
  screen_distance = screen_distance_slider.val*(10**-2)
  distance_between_slits = distance_between_slits_slider.val*(10**-3)
  number_of_slits = np.floor(number_of_slits_slider.val)
  Y1 = grated_diffraction_intensity(slit_width, wavelength1, screen_distance, distance_between_slits, number_of_slits, X)
  Y2 = grated_diffraction_intensity(slit_width, wavelength2, screen_distance, distance_between_slits, number_of_slits, X)
  Y3 = grated_diffraction_intensity(slit_width, wavelength3, screen_distance, distance_between_slits, number_of_slits, X)
  plot1.set_ydata(Y1)
  plot2.set_ydata(Y2)
  plot3.set_ydata(Y3)
  plot.set_ydata(Y1+Y2+Y3)

wavelength1_slider.on_changed(update)
wavelength2_slider.on_changed(update)
wavelength3_slider.on_changed(update)
slit_width_slider.on_changed(update)
screen_distance_slider.on_changed(update)
distance_between_slits_slider.on_changed(update)
number_of_slits_slider.on_changed(update)

plt.show()
```
